```python
# Generar una lista 1D con números aleatorios
import random

LISTA_1D = [random.randrange(1, 1000000) for x in range(1000)]

# Encontrar los dos puntos más cercanos usando Fuerza Bruta
def encontrar_puntos_mas_cercanos_1d(lista):
    min_distancia = float('inf')
    puntos_cercanos = (None, None)
    for i in range(len(lista)):
        for j in range(i + 1, len(lista)):
            distancia = abs(lista[i] - lista[j])
            if distancia < min_distancia:
                min_distancia = distancia
                puntos_cercanos = (lista[i], lista[j])
    return puntos_cercanos, min_distancia

# Calcular los puntos más cercanos en 1D y su distancia
puntos_cercanos_1d, distancia_1d = encontrar_puntos_mas_cercanos_1d(LISTA_1D)

# Mostrar los resultados
print("Puntos más cercanos en 1D:", puntos_cercanos_1d)
print("Distancia mínima en 1D:", distancia_1d)
```

### **Tiempo de Complejidad del Algoritmo de Fuerza Bruta**

El algoritmo de fuerza bruta para encontrar los dos puntos más cercanos en 1D compara todas las posibles parejas de puntos en la lista. Esto implica un número total de $ \binom{n}{2} $, es decir:

$$
\text{Comparaciones} = \frac{n \cdot (n-1)}{2}
$$

Por lo tanto, la complejidad temporal del algoritmo es $ O(n^2) $, donde $ n $ es el número de puntos en la lista. Esto ocurre porque se utilizan dos bucles anidados para calcular la distancia entre cada par de puntos.

---

### **Mejora con Divide y Vencerás**

El enfoque de **Divide y Vencerás** mejora la complejidad temporal al dividir la lista en subproblemas más pequeños:

1. **Ordenar los puntos en la lista** (complejidad $ O(n \log n) $).
2. **Dividir la lista ordenada** en dos mitades y resolver recursivamente el problema para cada mitad.
3. **Comparar los puntos más cercanos** de cada mitad, considerando también los puntos cercanos a la frontera entre ambas mitades.
4. **La combinación de las soluciones** tiene un coste lineal ($ O(n) $).

La complejidad total del algoritmo es $ O(n \log n) $, significativamente mejor que $ O(n^2) $.


```python
# Encontrar los puntos más cercanos usando Divide y Vencerás
def cercanos_dyv(lista):
    # Función auxiliar para encontrar el mínimo entre pares de puntos
    def closest_pair_rec(sublista):
        if len(sublista) <= 3:  # Resolver mediante fuerza bruta si hay pocos elementos
            puntos_cercanos, distancia = encontrar_puntos_mas_cercanos_1d(sublista)
            return puntos_cercanos[0], puntos_cercanos[1], distancia

        # Dividir la lista en dos mitades
        mid = len(sublista) // 2
        izquierda = sublista[:mid]
        derecha = sublista[mid:]

        # Resolver recursivamente
        p1, q1, d1 = closest_pair_rec(izquierda)
        p2, q2, d2 = closest_pair_rec(derecha)

        # Obtener el mínimo entre las soluciones izquierda y derecha
        min_d = min(d1, d2)
        if d1 < d2:
            min_pair = (p1, q1)
        else:
            min_pair = (p2, q2)

        # Considerar puntos cercanos a la división
        frontera = [x for x in sublista if abs(x - sublista[mid]) < min_d]
        frontera.sort()

        for i in range(len(frontera)):
            for j in range(i + 1, len(frontera)):
                if abs(frontera[j] - frontera[i]) >= min_d:
                    break
                distancia = abs(frontera[j] - frontera[i])
                if distancia < min_d:
                    min_d = distancia
                    min_pair = (frontera[i], frontera[j])

        return min_pair[0], min_pair[1], min_d

    # Ordenar la lista antes de llamar a la función recursiva
    lista.sort()
    p1, p2, distancia = closest_pair_rec(lista)
    return (p1, p2), distancia

# Aplicar el algoritmo al dataset anterior
puntos_cercanos_1d_divide, distancia_1d_divide = cercanos_dyv(LISTA_1D)

# Mostrar los resultados
print("Puntos más cercanos con Divide y Vencerás:", puntos_cercanos_1d_divide)
print("Distancia mínima con Divide y Vencerás:", distancia_1d_divide)
```

### Mejora divide y Vencerás

Cuando se trata de encontrar los dos valores más cercanos en una lista, existen varios algoritmos que podemos considerar. Divide y vencerás es muy eficiente y aunque no podamos reducir la complejidad de este algoritmo sí que podemos encontrar uno más rápido para encontrar la solución con un pequeño tratamiento sobre el conjunto de datos.

### **1. Divide y Vencerás**

#### **Cómo Funciona**
1. Ordena los puntos en la lista ($O(n \log n)$).
2. Divide la lista en dos mitades y resuelve recursivamente el problema para cada mitad.
3. Combina las soluciones, comparando puntos cercanos a la frontera entre las mitades.

### **2. Ordenar y Comparar Vecinos**

#### **Cómo Funciona**
1. Ordena la lista ($O(n \log n)$).
2. Compara los valores consecutivos en la lista ordenada ($O(n)$) para encontrar los puntos más cercanos.

#### **Ventajas**
- Simplicidad extrema en su implementación.
- Igual de rápido que Divide y Vencerás en términos de complejidad ($O(n \log n)$).
- Ideal para problemas en 1D.

#### **Desventajas**
- Solo es aplicable a problemas en 1D. En dimensiones más altas, no puede competir con Divide y Vencerás.

### **Conclusión**

Si se trabaja en problemas 1D, el enfoque de **Ordenar y Comparar Vecinos** es la mejor opción debido a su simplicidad y eficiencia. Sin embargo, si necesitas trabajar en 2D o dimensiones más altas, el algoritmo de **Divide y Vencerás** es el más adecuado, ya que puede manejar estos casos con eficiencia.



```python
# Encontrar los dos puntos más cercanos ordenando y comparando vecinos consecutivos
def cercanos_ordenados(lista):
    # Ordenar la lista
    lista.sort()

    # Comparar vecinos consecutivos
    min_distancia = float('inf')
    puntos_cercanos = (None, None)
    for i in range(len(lista) - 1):
        distancia = abs(lista[i] - lista[i + 1])
        if distancia < min_distancia:
            min_distancia = distancia
            puntos_cercanos = (lista[i], lista[i + 1])

    return puntos_cercanos, min_distancia

# Aplicar el algoritmo al dataset anterior
puntos_cercanos_ordenados, distancia_ordenados = cercanos_ordenados(LISTA_1D)

# Mostrar los resultados
print("Puntos más cercanos con Ordenar y Comparar Vecinos:", puntos_cercanos_ordenados)
print("Distancia mínima con Ordenar y Comparar Vecinos:", distancia_ordenados)
```

A continuación se implementa divide y vencerás en datos de 2 y 3 dimensiones.

```python
# Encontrar los puntos más cercanos en 2D usando Divide y Vencerás
def cercanos_dyv_2d(lista):
    # Función para calcular la distancia euclidiana
    def distancia_2d(p1, p2):
        return ((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)**0.5

    # Función auxiliar recursiva
    def closest_pair_rec(sublista_x, sublista_y):
        if len(sublista_x) <= 3:  # Fuerza bruta si pocos puntos
            min_distancia = float('inf')
            puntos_cercanos = (None, None)
            for i in range(len(sublista_x)):
                for j in range(i + 1, len(sublista_x)):
                    distancia = distancia_2d(sublista_x[i], sublista_x[j])
                    if distancia < min_distancia:
                        min_distancia = distancia
                        puntos_cercanos = (sublista_x[i], sublista_x[j])
            return puntos_cercanos, min_distancia

        # Dividir la lista en dos mitades
        mid = len(sublista_x) // 2
        mitad_x = sublista_x[mid][0]
        izquierda_x = sublista_x[:mid]
        derecha_x = sublista_x[mid:]
        izquierda_y = list(filter(lambda p: p[0] <= mitad_x, sublista_y))
        derecha_y = list(filter(lambda p: p[0] > mitad_x, sublista_y))

        # Resolver recursivamente
        (p1, q1), d1 = closest_pair_rec(izquierda_x, izquierda_y)
        (p2, q2), d2 = closest_pair_rec(derecha_x, derecha_y)

        # Seleccionar la menor distancia entre las dos mitades
        min_d = min(d1, d2)
        min_pair = (p1, q1) if d1 < d2 else (p2, q2)

        # Considerar puntos cercanos a la frontera
        frontera = [p for p in sublista_y if abs(p[0] - mitad_x) < min_d]
        for i in range(len(frontera)):
            for j in range(i + 1, min(i + 7, len(frontera))):  # Verificar hasta 7 vecinos
                distancia = distancia_2d(frontera[i], frontera[j])
                if distancia < min_d:
                    min_d = distancia
                    min_pair = (frontera[i], frontera[j])

        return min_pair, min_d

    # Ordenar la lista por coordenadas X y Y
    lista_x = sorted(lista, key=lambda p: p[0])
    lista_y = sorted(lista, key=lambda p: p[1])
    return closest_pair_rec(lista_x, lista_y)

# Generar una lista de puntos aleatorios en 2D
LISTA_2D = [(random.randint(1, 1000000), random.randint(1, 1000000)) for _ in range(1000)]

# Aplicar el algoritmo a los puntos en 2D
puntos_cercanos_2d, distancia_2d = cercanos_dyv_2d(LISTA_2D)
print("Puntos más cercanos en 2D con Divide y Vencerás:", puntos_cercanos_2d)
print("Distancia mínima en 2D con Divide y Vencerás:", distancia_2d)
```

```python
# Encontrar los puntos más cercanos en 3D usando Divide y Vencerás
def cercanos_dyv_3d(lista):
    # Función para calcular la distancia euclidiana
    def distancia_3d(p1, p2):
        return ((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)**0.5

    # Función auxiliar recursiva
    def closest_pair_rec(sublista_x, sublista_y, sublista_z):
        if len(sublista_x) <= 3:  # Fuerza bruta si pocos puntos
            min_distancia = float('inf')
            puntos_cercanos = (None, None)
            for i in range(len(sublista_x)):
                for j in range(i + 1, len(sublista_x)):
                    distancia = distancia_3d(sublista_x[i], sublista_x[j])
                    if distancia < min_distancia:
                        min_distancia = distancia
                        puntos_cercanos = (sublista_x[i], sublista_x[j])
            return puntos_cercanos, min_distancia

        # Dividir la lista en dos mitades
        mid = len(sublista_x) // 2
        mitad_x = sublista_x[mid][0]
        izquierda_x = sublista_x[:mid]
        derecha_x = sublista_x[mid:]
        izquierda_y = list(filter(lambda p: p[0] <= mitad_x, sublista_y))
        derecha_y = list(filter(lambda p: p[0] > mitad_x, sublista_y))
        izquierda_z = list(filter(lambda p: p[0] <= mitad_x, sublista_z))
        derecha_z = list(filter(lambda p: p[0] > mitad_x, sublista_z))

        # Resolver recursivamente
        (p1, q1), d1 = closest_pair_rec(izquierda_x, izquierda_y, izquierda_z)
        (p2, q2), d2 = closest_pair_rec(derecha_x, derecha_y, derecha_z)

        # Seleccionar la menor distancia entre las dos mitades
        min_d = min(d1, d2)
        min_pair = (p1, q1) if d1 < d2 else (p2, q2)

        # Considerar puntos cercanos a la frontera
        frontera = [p for p in sublista_y if abs(p[0] - mitad_x) < min_d]
        for i in range(len(frontera)):
            for j in range(i + 1, min(i + 7, len(frontera))):  # Verificar hasta 7 vecinos
                distancia = distancia_3d(frontera[i], frontera[j])
                if distancia < min_d:
                    min_d = distancia
                    min_pair = (frontera[i], frontera[j])

        return min_pair, min_d

    # Ordenar la lista por coordenadas X, Y y Z
    lista_x = sorted(lista, key=lambda p: p[0])
    lista_y = sorted(lista, key=lambda p: p[1])
    lista_z = sorted(lista, key=lambda p: p[2])
    return closest_pair_rec(lista_x, lista_y, lista_z)

# Generar una lista de puntos aleatorios en 3D
LISTA_3D = [(random.randint(1, 1000000), random.randint(1, 1000000), random.randint(1, 1000000)) for _ in range(1000)]

# Aplicar el algoritmo a los puntos en 3D
puntos_cercanos_3d, distancia_3d = cercanos_dyv_3d(LISTA_3D)
print("Puntos más cercanos en 3D con Divide y Vencerás:", puntos_cercanos_3d)
print("Distancia mínima en 3D con Divide y Vencerás:", distancia_3d)
```
